# üîπ Step 2.1: Load the Data 1

```python
df1=pd.read_csv(r"C:\Users\USER\Downloads\eda-crypto-master\datasets\coinmarketcap_06012018.csv")
df2=pd.read_csv(r"C:\Users\USER\Downloads\eda-crypto-master\datasets\coinmarketcap_06122017.csv")
df1.info()
df2.info()
```

# Step 2.2: Load the Data 2

```python
df2=pd.read_csv(r"C:\Users\USER\Downloads\eda-crypto-master\datasets\coinmarketcap_06122017.csv")
df2.info()
```

# üîπ Step 4.1: Handle Missing Values

```python
print(df1.isnull().sum())

# Filling missing values with zero
df1.fillna(0, inplace=True)
```

#  Step 4.2: Handle Missing Values

## Market Capitalization Distribution using Histoplot

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.histplot(df1['market_cap_usd'], bins=50, kde=True, color='blue', label='Jan 2018')
sns.histplot(df2['market_cap_usd'], bins=50, kde=True, color='red', label='Dec 2017')
plt.xscale("log")
plt.legend()
plt.title("Market Capitalization Distribution")
plt.xlabel("Market Cap (USD)")
plt.ylabel("Frequency")
plt.show()
```

# Top 10 Cryptocurrencies by Market Cap (Jan 2018) using Barplot

```python
top10_2018 = df1.nlargest(10, 'market_cap_usd')

plt.figure(figsize=(12, 6))
sns.barplot(y=top10_2018['name'],hue=top10_2018['name'], x=top10_2018['market_cap_usd'], palette='Blues_r')
plt.xlabel("Market Cap (USD)")
plt.ylabel("Cryptocurrency")
plt.title("Top 10 Cryptocurrencies by Market Cap (Jan 2018)")
plt.show()
```

# Top 10 Cryptocurrencies by Market Cap (Jan 2017) using Barplot

```python
top10_2018 = df2.nlargest(10, 'market_cap_usd')

plt.figure(figsize=(12, 6))
sns.barplot(y=top10_2018['name'],hue=top10_2018['name'], x=top10_2018['market_cap_usd'], palette='Blues_r')
plt.xlabel("Market Cap (USD)")
plt.ylabel("Cryptocurrency")
plt.title("Top 10 Cryptocurrencies by Market Cap (Jan 2017)")
plt.show()
```

#  Price vs Market Cap (Jan 2018) using Scatterplot

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df1, x='market_cap_usd', y='price_usd', alpha=0.5)
plt.xscale("log")
plt.yscale("log")
plt.title("Price vs Market Cap (Jan 2018)")
plt.xlabel("Market Cap (USD)")
plt.ylabel("Price (USD)")
plt.show()
```

# Feature Correlation (Jan 2018) using Heatmap

```python
# Select only numeric columns
numeric_df1 = df1.select_dtypes(include=['number'])

# Generate correlation heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df1.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation (Jan 2018)")
plt.show()
```

#   Feature Correlation (Jan 2017) using Heatmap

```python
# Select only numeric columns
numeric_df2 = df2.select_dtypes(include=['number'])

# Generate correlation heatmap
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df2.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Feature Correlation (Jan 2017)")
plt.show()
```

# Price Distribution using Histplot

```python
plt.figure(figsize=(12, 6))
sns.histplot(df1['price_usd'], bins=50, kde=True, color='green', label='Jan 2018')
sns.histplot(df2['price_usd'], bins=50, kde=True, color='purple', label='Dec 2017')
plt.xscale("log")
plt.legend()
plt.title("Price Distribution")
plt.xlabel("Price (USD)")
plt.ylabel("Frequency")
plt.show()
```

#  24h Percentage Change Distribution (Jan 2017) using Boxplot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=df2, x='percent_change_24h', color='blue')
plt.title("24h Percentage Change Distribution (Jan 2017)")
plt.xlabel("Percentage Change")
plt.show()
```

#  Total Supply vs Market Cap (Jan 2018) using Scatterplot

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df1, x='total_supply', y='market_cap_usd', alpha=0.5)
plt.xscale("log")
plt.yscale("log")
plt.title("Total Supply vs Market Cap (Jan 2018)")
plt.xlabel("Total Supply")
plt.ylabel("Market Cap (USD)")
plt.show()
```
